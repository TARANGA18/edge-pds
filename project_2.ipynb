```python
df_tt1 = pd.read_csv('term-test-1-result.csv')
df_tt2 = pd.read_csv('term-test-2-result.csv')

df_merged = pd.merge(df_tt1, df_tt2, on='Registration Number')

print(df_merged.head())
```

```python
df_merged['Best Marks'] = df_merged[['TT-1 Marks', 'TT-2 Marks']].max(axis=1)

df_merged['Average Marks'] = df_merged[['TT-1 Marks', 'TT-2 Marks']].mean(axis=1)

print(df_merged.head())
```

```python
df_final = df_merged.drop(['TT-1 Marks', 'TT-2 Marks'], axis=1)

print(df_final.head())
```

```python
df_final.to_csv('final-term-test-result.csv', index=False)
print(df_final.head())
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X = df_final[['Average Marks']]

kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

df_final['Cluster'] = kmeans.labels_

plt.scatter(df_final.index, df_final['Average Marks'], c=kmeans.labels_, cmap='viridis', marker='o')
plt.xlabel('Student Index')
plt.ylabel('Average Marks')
plt.title('K-Means Clustering of Average Marks')
plt.colorbar()
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the final DataFrame
df_final = pd.read_csv('final-term-test-result.csv')

# Extract the average marks for clustering
X = df_final[['Average Marks']]

# Perform K-Means clustering for a range of clusters (k)
inertia = []
for k in range(1, 6):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Plot the elbow curve to visualize the optimal number of clusters
plt.plot(range(1, 6), inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.show()
```
